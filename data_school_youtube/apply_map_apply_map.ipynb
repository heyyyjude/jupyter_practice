```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

### Columns 추가

```python
ufo["Location"] = ufo["City"] + ", " + ufo["State"]
ufo.head()
```

### Rename columns by dictionary

```python
ufo.rename(columns={'Colors Reported':'Colors_Reported', 'Shape Reported':'Shape_Reported'}, inplace=True)
ufo.head()
```

ufo.columns=["City", "Colors_Reported", "Shape_Reported", "State", "Time", "Location"]
ufo.head()

### inplace 현재 데이터 프레임을 수정한다. 그냥 drop하면 현재 데이터 프레임에는 반영되지 않음.

```python
tmp = ufo

tmp.drop("Colors_Reported", axis=1)

tmp.drop("Colors_Reported", axis=1, inplace=True)
```

```python
ufo.drop(["City", "State"], axis=1).head()
```

# 9. Multiple filter criteria to DF

```python
movies.head()
```

```python
movies[(movies.duration>=200) & (movies.genre=="Drama")].head()
```

```python
movies[(movies.duration>=200) | (movies.genre=="Drame")].head()
```

```python
movies[movies.genre.isin(["Crime", "Drame", "Action"])].head()
```
