```python
ages = [0,2,10,21,23,37,31,61,20,41,32]
bins = [1,15,25,35,60,190]
labels = ["미성년자", "청년", "중년", "장년", "노년"]

cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
cats.codes
```

```python
df = pd.DataFrame(ages, columns=["ages"])
df["age_cat"] = pd.cut(df.ages, bins, labels=labels)
df
```
